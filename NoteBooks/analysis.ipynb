```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/superstore.csv')
print("Initial Data Loaded.")
```

```python
df.columns = [c.replace('.', '_') for c in df.columns]
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
df.to_csv('cleaned_superstore_for_tableau.csv', index=False)
print("SUCCESS: Your cleaned table is ready for Tableau!")
```

```python
df = pd.read_csv('cleaned_superstore_for_tableau.csv')
df['Order_Date'] = pd.to_datetime(df['Order_Date'])
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='Category', y='Profit', data=df, estimator=sum, hue='Category', palette='viridis', legend=False)
plt.title('Total Profit by Category')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
df.resample('ME', on='Order_Date')['Sales'].sum().plot(kind='line', marker='o', color='teal')
plt.title('Monthly Sales Trend')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```
